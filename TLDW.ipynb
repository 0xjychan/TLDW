#### Convert audio to text

```python
import whisper
```

```python
# Load Whisper model
model = whisper.load_model('large')
```

```python
# Transcribe audio of YouTube video
output_text = model.transcribe('./audio/yt_audio1.mp3', language='english')
```

```python
output_text['text']
```

#### Summarize text with Pegasus-xsum


```python
from transformers import PegasusForConditionalGeneration, PegasusTokenizer
```

```python
# Load tokenizer
tokenizer = PegasusTokenizer.from_pretrained("google/pegasus-xsum")
```

```python
# Load model
summary_model = PegasusForConditionalGeneration.from_pretrained("google/pegasus-xsum")
```

```python
# Create tokens
tokens = tokenizer(output_text['text'], truncation=True, padding='longest', return_tensors='pt')
```

```python
# Summarize
summary = summary_model.generate(**tokens)
```

```python
# Decode summary
tokenizer.decode(summary[0])
```
